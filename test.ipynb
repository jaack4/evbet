```python
import matplotlib.pyplot as plt

# Filter for overs only
overs_df = josh_allen_df[josh_allen_df['outcome'] == 'Over'].copy()

# Get unique markets
markets = overs_df['market'].unique()

# Create subplots for each market
fig, axes = plt.subplots(len(markets), 1, figsize=(12, 5*len(markets)))
if len(markets) == 1:
    axes = [axes]

for idx, market in enumerate(markets):
    market_data = overs_df[overs_df['market'] == market]
    
    # Plot each bookmaker with different color/marker
    for bookmaker in market_data['bookmaker'].unique():
        bm_data = market_data[market_data['bookmaker'] == bookmaker]
        axes[idx].scatter(bm_data['line'], bm_data['price'], label=bookmaker, s=100, alpha=0.7)
    
    axes[idx].set_xlabel('Line', fontsize=12)
    axes[idx].set_ylabel('Price (Decimal Odds)', fontsize=12)
    axes[idx].set_yscale('log')
    axes[idx].set_title(f'{market} - Overs (Josh Allen)', fontsize=14, fontweight='bold')
    axes[idx].legend(loc='best')
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
